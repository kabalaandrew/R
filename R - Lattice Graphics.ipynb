```r
require(lattice)
xyplot(mpg ~ wt, data=mtcars)
```

## Univariate Lattice Graphics

```r
histogram( ~ mpg, data=mtcars)
```

```r
densityplot( ~ mpg, data=mtcars)
```

```r
bwplot(mpg ~ carb, data=mtcars, horizontal=FALSE)
```

## Scatter plot matrix "Splom" and "pairs"

```r
splom( ~ mtcars[, c("mpg", "wt", "cyl", "hp")])
```

```r
xyplot(mpg ~ wt, data=mtcars, main="Miles per Gallon vs Weight", xlab="Weight (lb/1000)",
      ylab="Miles/(US) Gallon", xlim=c(0,6), ylim=c(10,40))
```

```r
library(cranlogs)
cranData <- cran_downloads(packages=c("lattice","ggplot2"), when="last-month")
head(cranData)
```

```r
xyplot(count ~ date, data=cranData, subset=package=="lattice", main="Lattice package downloads over the last month",
      ylab="Number of downloads", xlab="Date", type="b", col="red", lwd=2, cex=2, pch=16)
```

## Multiple variables

```r
xyplot(mpg ~ disp + hp, data=mtcars, auto.key = TRUE, pch=16, cex = 2)
```

```r
xyplot(mpg ~ disp + hp, data=mtcars, auto.key = TRUE, pch=16, cex = 2, outer = TRUE)
```

## Groups of data

```r
xyplot(mpg ~ wt, data=mtcars, groups=cyl, pch=16, cex=2, auto.key=TRUE)
```

```r
xyplot(mpg ~ disp + hp, data=mtcars, groups=cyl, pch=16, cex = 2, auto.key=TRUE)
```

## Using Panels

```r
xyplot(count ~ date | package, data=cranData, type="b", col="red")
```

## Controlling strip headers

```r
xyplot(mpg~wt | cyl, data=mtcars, main="Miles per Gallon vs Weight by Number of Cylinders")
```

```r
xyplot(mpg~wt | factor(cyl), data=mtcars, main="Miles per Gallon vs Weight by Number of Cylinders")
```

## Multiple By Variables

```r
# cars auto am==0 cars,  manual am==1
xyplot(mpg ~ wt | factor(cyl) * ifelse(am==0, "Automatic", "Manual"), data=mtcars, cex=1.5, pch=21, fill="lightblue", 
      Main="Miles per Gallon vs Weight\nBy Number of cylinders and Transmission type")
```
