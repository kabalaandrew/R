```r
library(tidyverse)
library(viridis)
library(RColorBrewer)
```

```r
set.seed(12345)
```

```r
# create a dataframe
df.data <- tibble(x=runif(20000), y=rnorm(20000))
head(df.data)
```

```r
# create a scatterplot
ggplot(df.data, aes(x=x, y=y)) + geom_point()
```

```r
# Density plot with default fill
ggplot(df.data, aes(x=x,y=y)) + stat_density2d(aes(fill=..density..), geom="tile",
                                              contour=F)
```

```r
#Display color palettes
display.brewer.all()
```

```r
#--------------------------------
# FILL WITH COLOR PALETTE: Greens
#--------------------------------

ggplot(df.data, aes(x = x, y = y)) +
  stat_density2d(aes(fill = ..density..), geom = 'tile', contour = F) +
  scale_fill_distiller(palette = 'Greens')
```

```r
#---------------------------------
# FILL WITH COLOR PALETTE: Viridis
#---------------------------------

ggplot(df.data, aes(x = x, y = y)) +
  stat_density2d(aes(fill = ..density..), geom = 'tile', contour = F) +
  scale_fill_viridis()
```

## LINE CHARTS IN R

```r
library(stringr)
# using readr to read the CSV
stock_amzn <- read_csv("http://sharpsightlabs.com/wp-content/uploads/2017/09/AMZN_stock.csv")
```

```r
# inspect
stock_amzn %>% names()
stock_amzn %>% head()
```

```r
#=========================================================
# CHANGE COLUMN NAMES: lower case
# - in the raw form (as read in) the first letter of 
#   each variable is capitalized. 
# - This makes them harder to type!  Not ideal.
# - we'll use stringr::str_to_lower() to change the column
#   names to lower case 
#=========================================================

colnames(stock_amzn) <- colnames(stock_amzn) %>% str_to_lower()

# inspect
stock_amzn %>% names()
```

```r
ggplot(stock_amzn, aes(x=date, y=close)) + geom_line()
```

```r
ggplot(stock_amzn, aes(x=date, y=close)) + 
geom_line(color='cyan') + 
geom_area(fill='cyan', alpha=.1) + 
labs(x='Date', y='Closing\nPrice',
    title="Amazon's stock price has increased dramatically\nOver the last 20 Years") +
theme(text = element_text(family = 'Gill Sans', color = "#444444")
        ,panel.background = element_rect(fill = '#444B5A')
        ,panel.grid.minor = element_line(color = '#4d5566')
        ,panel.grid.major = element_line(color = '#586174')
        ,plot.title = element_text(size = 28)
        ,axis.title = element_text(size = 18, color = '#555555')
        ,axis.title.y = element_text(vjust = 1, angle = 0)
        ,axis.title.x = element_text(hjust = 0)
        ) 
```
